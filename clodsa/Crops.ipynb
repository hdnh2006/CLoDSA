```python
from matplotlib import pyplot as plt
from clodsa.augmentors.augmentorFactory import createAugmentor
from clodsa.augmentors.generator import Generator
from clodsa.techniques.techniqueFactory import createTechnique
import cv2
%matplotlib inline
```

```python
PROBLEM = "detection"
```

```python
ANNOTATION_MODE = "pascalvoc"
```

```python
INPUT_PATH = "../anotaciones/"
```

```python
GENERATION_MODE = "linear"
```

```python
OUTPUT_MODE = "pascalvoc"
OUTPUT_PATH= "../anotaciones-crop/"
```

```python
augmentor = createAugmentor(PROBLEM,ANNOTATION_MODE,OUTPUT_MODE,GENERATION_MODE,INPUT_PATH,{"outputPath":OUTPUT_PATH})
```

```python
PATCH_WIDTH,PATCH_HEIGHT,STEP_WIDTH,STEP_HEIGHT = 400,400,400,400
```

```python
MAX_WIDTH,MAX_HEIGHT=3000,3000
```

```python
startY = 0
while(startY<MAX_HEIGHT):
    startX = 0
    while(startX<MAX_WIDTH):
        crop=createTechnique("cropSize", {"x" : startX,"y":startY,"width":PATCH_WIDTH,"height":PATCH_HEIGHT})
        augmentor.addGenerator(Generator(crop))
        startX += STEP_WIDTH
    startY += STEP_HEIGHT
```

```python
augmentor.applyAugmentation()
```
